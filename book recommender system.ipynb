POPULARITY BASED RECOMMENDER SYSTEM


```python
ratings_with_name = ratings.merge(books,on='ISBN') #THIS WILL MERGE BOTH THE RATINGS DATASET  AND BOOKS DATASET using the attribute which is common in both the tables ISBN value 
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df 
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df #merging the total number of ratings given by a person with that of average rating of that particular book
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

COLLABRATIVVE FILTERING BASED RECOMMENDER SYSTEM

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]#this will check for those users which matches the padhe likhe users and the changes made were stored in the filtered_rating df 
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index 
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)# this will fill all the emply or NA  values that are present in the dataset into 0 and all the changes will be done in our original dataset only
```

```python
pt #this will consists of all the user id and book title in discriptive form becuase of the use of pivot table
```

```python
from sklearn.metrics.pairwise import cosine_similarity ## cosine similarity will be used to find the euclidian distance between the 2 books and their similarity so that it could suggest similar books
```

```python

similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape ## this will find the similarity between the books thats why pt is passed
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
